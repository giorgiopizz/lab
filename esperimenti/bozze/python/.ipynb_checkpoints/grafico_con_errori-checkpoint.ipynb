## Grafici con errore e definizione di una funzione per il fit
Per definire una funzione da fittare usare il prototipo def f(x,parametri) con x e parametri due array. Per usare funzioni come sin,cos,exp, eccetera è meglio usare quelle definite da ROOT in TMath. 
### RICORDA: i dati per il grafico devono essere ordinati nella x
In automatico il grafico connette punti consecutivo nell'ordine di lettura dal file perciò verrebbe un pasticcio se ci fossero punti registrati in modo disordinato

```python
import ROOT
#import TMath
import math
import random
def funzione(x,parametri):
    return parametri[0]*ROOT.TMath.Exp(parametri[1]*x[0])
try:
    del cnv
    cnv=ROOT.TCanvas("Canvas","Canvas",2000,1200)
except NameError:
    cnv=ROOT.TCanvas("Canvas","Canvas",2000,1200)
graph=ROOT.TGraphErrors("dati.txt")
f=ROOT.TF1("func",funzione,0,10,2)
f.SetParameter(0,3)
f.SetParameter(1,-3)
r=graph.Fit(f,"s")
graph.Draw("APE")
leg=ROOT.TLegend(.1,.7,.3,.9,"Lab. Lesson 1")
leg.SetFillColor(0)
graph.SetFillColor(0)
leg.AddEntry(graph,"Exp. Points")
leg.AddEntry(f,"Th. Law")
leg.Draw("Same")
cnv.Draw()
```

```python
r.Print()
cov=r.GetCovarianceMatrix()
cov.Print()
```

## Metodo per la creazione di dati in python usando TRandom3(uno dei migliori di ROOT)

```python
import random, math,numpy as np
#import ROOT
rnd=ROOT.TRandom3()
def func(x):
    y=10*math.exp(-2*x)
    return y+rnd.Gaus(-y/100,y/100)
with open("dati.txt","w") as f:
    array=list(np.linspace(0,2,100))
    for i in range(len(array)):
        x=array[i]
        y=func(x)
        xerr=(10-0)/1000
        yerr=y/10
        stringa="%f %f %f %f\n"%(x,y,xerr,yerr)
        f.write(stringa)
```
