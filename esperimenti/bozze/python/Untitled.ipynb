```python
import ROOT
#import TMath
import math
import random
def funzione(x,parametri):
    return parametri[0]*ROOT.TMath.Exp(parametri[1]*x[0])
cnv=ROOT.TCanvas("Canvas","Canvas",800,500)
graph=ROOT.TGraphErrors("dati.txt")
f=ROOT.TF1("func",funzione,-10,10,2)
graph.Fit(f)
graph.Draw()
cnv.Draw()
```

```python
import random, math,numpy as np
def func(x):
    return 3*math.exp(-2*x)
with open("dati.txt","w") as f:
    array=list(np.linspace(-10,10))
    for i in range(len(array)):
        x=array[i]
        y=func(x)
        xerr=0.2
        yerr=y/10
        stringa="%f %f %f %f\n"%(x,y,xerr,yerr)
        f.write(stringa)
```

```python
def append(x,y,xerr,yerr,a,b,c,d): 
    x.append(a) 
    y.append(b) 
    xerr.append(c) 
    yerr.append(d) 
    return x,y,xerr,yerr
def read_errors(file):
    x=[];y=[];xerr=[];yerr=[]
    with open(file) as f: 
        line=f.readline() 
        i=1 
        while line:
            a,b,c,d=line.split() 
            x,y,xerr,yerr=append(x,y,xerr,yerr,a,b,c,d)
            line=f.readline() 
            i+=1 
    return i-1,x,y,xerr,yerr
from matplotlib import pyplot as plt

#TGraphErrors(read_errors("dati.txt"))
i,x,y,xerr,yerr=read_errors("dati.txt")
x=list(map(float,x))
y=list(map(float,y))
xerr=list(map(float,xerr))
yerr=list(map(float,yerr))
plt.errorbar(x,y,yerr=0.5,xerr=0.001)
```
